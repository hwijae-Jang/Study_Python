## 4. 점수가 90이상이면 A, 80 >= B , 70 <= C 등급 측정

```python
df = pd.DataFrame([[100, 'C100', '컴퓨터구조', 91],
                     [200, 'C200', '데이터베이스', 85],
                     [400, 'C300', '운영체제', 65],
                     [200, 'C300', '운영체제', 85],
                     [300, 'C100', '컴퓨터구조', 100],
                     [500, 'C200', '데이터베이스', 55]],
                     index = range(1,7), columns=['학번','과목번호','과목이름','점수'])
df 
```

```python
df['등급'] = np.where(df['점수'] >= 90,"A",np.where(df['점수']>=80,"B","C")) 

df['등급']

df
```

```python
grade = []
for i in df['점수']:
    if i >= 90: grade.append("A")
    elif i >= 80: grade.append("B")
    elif i <= 70: grade.append("C")
df['등급'] = grade 

df
```

```python
df['등급']= ["A" if i > 90 else "B" if i > 80 else 'C' for i in df['점수'] ]
df 
```
